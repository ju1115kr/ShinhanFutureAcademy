## 데이터 시각화
데이터 시각화는 데이터 분석 과정에서 중요한 단계 중 하나로, 데이터를 그래픽 형태로 표현하는 과정을 의미합니다. 이는 숫자나 텍스트로만 이루어진 복잡한 데이터 세트를 이해하기 쉽고, 직관적인 시각적 형태로 전환하여, 패턴, 추세, 이상치 같은 중요 정보를 빠르게 식별할 수 있게 해줍니다.

````python
import matplotlib.pyplot as plt # Python 시각화 기능 기반
import seaborn as sns # 자동으로 예쁜 그림, (연습용데이터)
import pandas as pd
````

````python
# 시각화방법 + 왜 시각화를 해야하는지?
# anscombe = pd.read_csv("Data/anscombe.csv")
anscombe = sns.load_dataset("anscombe")

data1 = anscombe[anscombe['dataset'] == 'I']
data2 = anscombe[anscombe['dataset'] == 'II']
data3 = anscombe[anscombe['dataset'] == 'III']
data4 = anscombe[anscombe['dataset'] == 'IV']
````

````python
# 표준편차
print(data1['x'].mean(), data1['y'].mean())
print(data2['x'].mean(), data2['y'].mean())
print(data3['x'].mean(), data3['y'].mean())
print(data4['x'].mean(), data4['y'].mean())
````

````python
print(data1['x'].std(), data1['y'].std())
print(data2['x'].std(), data2['y'].std())
print(data3['x'].std(), data3['y'].std())
print(data4['x'].std(), data4['y'].std())
````

````python
print(data1['x'].corr(data1['y']))
print(data2['x'].corr(data2['y']))
print(data3['x'].corr(data3['y']))
print(data4['x'].corr(data4['y']))
````

````python
fig = plt.figure() # 도화지

ax1 = fig.add_subplot(2, 2, 1) # 2행 2열의 1번째 자리
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.scatter(data1['x'], data1['y']) # 점 그래프 (산점도)
ax2.scatter(data2['x'], data2['y'])
ax3.scatter(data3['x'], data3['y'])
ax4.scatter(data4['x'], data4['y'])

ax1.set_title("data1")
ax2.set_title("data2")
ax3.set_title("data3")
ax4.set_title("data4")

fig.suptitle("Anscombe Data")

plt.tight_layout()
plt.savefig("Anscombe Data.png")
````

````python
# -, --, :, -.
plt.plot([1,2,3,4], [1,4,9,16], label = "Price", linestyle = "-") # 선 그래프 (자동으로 x축은 0부터 시작)
plt.plot([1,2,3,4], [2,5,3,10], label = "Demand", linestyle = "-.")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.legend() # 범례
plt.title("Title")
plt.xlim(0, 5)
plt.ylim(0, 20)
plt.show()
````

### 선 스타일
* ```-``` (Solid)
* ```--``` (Dashed)
* ```:``` (Dotted)
* ```-.``` (Dash-dot)

위 네가지 종류를 선택할 수 있습니다.

````python
plt.plot([1,2,3,4], [2,3,5,10], "r*--")
plt.plot([1,2,3,4], [3,4,6,11], "b^:") # 색깔, 마커, 선스타일
plt.grid(axis = "y")
plt.show()
````

````python
# 두 그래프 합쳐서 그리기

x = [2020, 2021, 2022, 2023, 2024, 2025, 2026]

y1 = [1, 3, 7, 5, 9, 7, 14]
y2 = [1, 3, 5, 7, 9, 11, 13]

fig, ax1 = plt.subplots() # 도화지, ax1이라는 자리 하나 할당.

ax1.plot(x, y1, 'gs-', linewidth = 5)

ax2 = ax1.twinx() # x축을 공유한다.
ax2.bar(x, y2, color = 'deeppink', alpha = 0.5)

ax1.set_zorder(ax2.get_zorder() + 10) # 위치값
ax1.patch.set_visible(False)
plt.show()
````

````python
import seaborn as sns

tips = pd.read_csv("Data/tips.csv")

# 연속형데이터 + 1개의 특성 : 히스토그램
sns.set_style("whitegrid")
sns.histplot(data=tips, x = 'total_bill', bins = 20, kde=True)
````

````python
# 2개의 연속형 데이터 : 산점도 (이변량 그래프)
sns.lmplot(data=tips, x = 'total_bill', y = 'tip', hue = "sex")
````

````python
import warnings
warnings.filterwarnings("ignore")

sns.lmplot(data=tips, x = 'total_bill', y = 'tip', hue = 'sex', fit_reg = False)
sns.kdeplot(data=tips, x = 'total_bill', y = 'tip')
````

````python
# 히스토그램 + 산점도
sns.jointplot(data=tips, x = 'total_bill', y = 'tip', kind = 'hex')
````

````python
sns.boxplot(data=tips, x = 'sex', y = 'tip', hue = 'day')
````

````python
# 산점도 행렬 (비교할 수 있는 모든 열을 비교해서 산점도)
sns.pairplot(tips, hue = 'sex')
````

````python
sns.violinplot(data=tips, x = 'sex', y = 'tip')
sns.swarmplot(data=tips, x = 'sex', y = 'tip', color = 'black')
````
