```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

samsung = pd.read_excel("samsung.xlsx")

X = samsung.
Y = samsung['Close']
```

```python
def EMA(data, col, span=2):
    return data[col].ewm(span=span).mean()

samsung['EMA(5)'] = EMA(samsung, 'Close', 5)
samsung['EMA(10)'] = EMA(samsung, 'Close', 10)
samsung['EMA(20)'] = EMA(samsung, 'Close', 20)
samsung['EMA(60)'] = EMA(samsung, 'Close', 60)
samsung['EMA(120)'] = EMA(samsung, 'Close', 120)

def Fourier(data, col, topn=2):
    import numpy as np
    fft = np.fft.fft(data[col])
    fft[topn:-topn] = 0
    ifft = np.fft.ifft(fft)
    return ifft
    
samsung['FFT(30)'] = Fourier(samsung, 'Close', 30)
```

```python
import seaborn as sns
samsung['EMA(120)'].plot()
samsung['FFT(30)'].plot()
```
