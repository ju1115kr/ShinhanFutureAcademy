```python
# 20회 정도 로또 정보 가져오기
total = []
for n in range(1, 21):
    url = requests.get(f"https://search.daum.net/search?w=tot&rtmaxcoll=LOT&DA=LOT&q={n}%ED%9A%8C%20%EB%A1%9C%EB%98%90%EB%8B%B9%EC%B2%A8%EB%B2%88%ED%98%B8")
    html = BeautifulSoup(url.text, 'html.parser')
    time.sleep(random.random() + 1) # 0~1 사이의 랜덤한값 +1 = 1.x초
    
    # 전체 번호 가져오기
    numbers = html.find("div", attrs = {'class': 'lottonum'}).find_all('span')
    del numbers[-3:-1]
    
    # 각 번째 번호마다 box에 넣기
    box = []
    for i in numbers:
        box.append(int(i.text))
    # 전체 번호는 total에 리스트로 넣기
    total. append(box)
        
    print(f"{n}회차 로또 저장완료 : {box}")
```

```python
###########################
# 네이버 주식 정보 가져오기(코스피)
###########################
import warnings
warnings.filterwarnings('ignore')

url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1") # 우선 한번 코스피(sosok=0)의 페이지를 한번 불러봄
html = BeautifulSoup(url.text)
kospi_page = int(html.find("td", class_ = "pgRR").find("a")['href'][-2:]) # 맨 뒷페이지의 페이지 번호는?

kospi_box = []
for page in range(1, kospi_page+1):
    url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={page}")
    html = BeautifulSoup(url.text)
    time.sleep(random.random() + 1) # 0~1 사이의 랜덤한값 +1 = 1.x초
    
    table = html.find('table', class_ = "type_2") # HTML 내 Table 정보를 가져옴
    table = pd.read_html(str(table))[0] # 이 테이블을 바로 pd로 전환하여 표현
    table = table.dropna(subset="종목명") # 이 중 종목명이 NA인 경우 제외(drop)
    
    del table['N']
    del table['토론실']
    #table = table.iloc[1:-3]
    
    kospi_box.append(table)
    print(f"코스피 {page}페이지 수집 완료")
    
kospi = pd.concat(kospi_box, ignore_index=True)
#kospi.to_excel("kospi.xlsx", index=False)
```

```python
###########################
# 네이버 주식 정보 가져오기(코스닥)
###########################
import warnings
warnings.filterwarnings('ignore')

url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page=1") # 우선 한번 코스닥(sosok=1)의 페이지를 한번 불러봄
html = BeautifulSoup(url.text)
kosdaq_page = int(html.find("td", class_ = "pgRR").find("a")['href'][-2:]) # 맨 뒷페이지의 페이지 번호는?

kosdaq_box = []
for page in range(1, kosdaq_page+1):
    url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page={page}")
    html = BeautifulSoup(url.text)
    time.sleep(random.random() + 1) # 0~1 사이의 랜덤한값 +1 = 1.x초
    
    table = html.find('table', class_ = "type_2") # HTML 내 Table 정보를 가져옴
    table = pd.read_html(str(table))[0] # 이 테이블을 바로 pd로 전환하여 표현
    table = table.dropna(subset="종목명") # 이 중 종목명이 NA인 경우 제외(drop)
    
    del table['N']
    del table['토론실']
    #table = table.iloc[1:-3]
    
    kosdaq_box.append(table)
    print(f"코스닥 {page}페이지 수집 완료")
    
kosdaq = pd.concat(kosdaq_box, ignore_index=True)
kosdaq.to_excel("kosdaq.xlsx", index=False)
```

```python
#####################
# 코스피/코스닥 소속 구분하고
# 코스피 + 코스닥(한국시장)
#####################
kospi['소속'] = '코스피'
kosdaq['소속'] = '코스닥'

korea_market = pd.concat([kospi, kosdaq], ignore_index=True)
korea_market.to_excel("korea_market.xlsx", index=False)
korea_market
```
