```python
# 이주영: PBR, PER, ROE별 가중치 필요하지 않나
# 김정수: PBR, PER이 산업군별로 상이 / 업종이 없음(자료가 부족)
# 기여은: 영업이익 순으로 산정필요
# 최진호: 자료가 부족하기에 배당률이 높거나 하는 등 특정 기준 하에서 진행하는 것이 필요하다고 봄. 미래 성장가능성에 투자하는 것이기에 
# 주식: 거래량이 높을수록 좋다. ROE가 높을수록 좋다(한 주당 얼마나 가치를 지니는 지), 매출액 증가추이, 영업이익 증가추이,
#      매출액 대비 영업이익, 유보율(주주배당하지않고, 유보하여 유보만큼 재투자)
# 분석대상: 거래량, 영업이익/매출액, ROE,매출액증가율, 영업이익증가율, 유보율

def remove_out(dataframe, remove_col):
    dff = dataframe
    for k in remove_col:
        level_1q = dff[k].quantile(0.25)
        level_3q = dff[k].quantile(0.75)
        IQR = level_3q - level_1q
        rev_range = 3  # 제거 범위 조절 변수
        dff = dff[(dff[k] <= level_3q + (rev_range * IQR)) & (dff[k] >= level_1q - (rev_range * IQR))]
        dff = dff.reset_index(drop=True)
    return dff

Good_PBR = stock["PBR"] <= 1
Good_PER = stock["PER"] <= 1
Good_ROE = stock["ROE"] >= 10
Good = stock[Good_PBR & Good_PER & Good_ROE]
```

```python
# 각 항목별 가중치를 두어, 필요한 데이터를 산정 후 재계산 
# 가중치항목: 영업이익증가율, 매출액증가율, 영업대비매출액(영업이익률/매출액), ROE, 거래량
# 이상치 제거함수(detect_outliers)를 두고 이상치 내용들은 제거함

import pandas as pd
stock = pd.read_excel("Data/2024-03-22_국내주식.xlsx")

def detect_outliers(df, columns):
    q1 = df[columns].quantile(0.25)
    q3 = df[columns].quantile(0.75)
    IQR = q3 - q1

    boundary = 1.5 * IQR

    index1 = df[df[columns] > q3 + boundary].index
    index2 = df[df[columns] < q1 - boundary].index

    df[columns] = df[columns].drop(index1)
    df[columns] = df[columns].drop(index2)

    return df

realStock = detect_outliers(stock, "PBR")
realStock = detect_outliers(realStock, "PER")
realStock = detect_outliers(realStock, "ROE")
realStock = detect_outliers(realStock, "ROA")
realStock = detect_outliers(realStock, "유보율")
realStock = detect_outliers(realStock, "외국인비율")
realStock = detect_outliers(realStock, "거래량")
realStock = detect_outliers(realStock, "거래대금")
realStock = detect_outliers(realStock, "시가총액")

realStock['영업대비매출액'] = realStock['영업이익']/realStock['매출액']
realStock['부채비율'] = realStock['부채총계'] / realStock['자산총계']

realStock = detect_outliers(realStock, "영업대비매출액")
realStock = detect_outliers(realStock, "부채비율")

realStock['R영업이익증가율'] = realStock['영업이익증가율'].rank(method='average', ascending = 1)
realStock['R매출액증가율'] = realStock['매출액증가율'].rank(method='average', ascending = 1)
realStock['R영업대비매출액'] = realStock['영업대비매출액'].rank(method='average', ascending = 1)
realStock['R부채비율'] = realStock['부채비율'].rank(method='average', ascending=0)
realStock['R_ROE'] = realStock['ROE'].rank(method='average', ascending = 1)
realStock['R거래량'] = realStock['거래량'].rank(method='average', ascending = 1)

realStock['score'] = realStock['R영업이익증가율'] * 0.25 + realStock['R매출액증가율'] * 0.2 + realStock['R영업대비매출액'] * 0.3 + realStock['R_ROE'] * 0.15 +   realStock['R거래량'] * 0.1
dropedStock = realStock.dropna()

dropedStock.sort_values('score', ascending=0)
```
