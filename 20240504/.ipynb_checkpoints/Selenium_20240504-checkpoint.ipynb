```python
#######################
# 인공신경망 원리
# 미분학습: 경사하강법
#######################

# 공부시간 -> 성적에 영향을 미치는가?
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([2, 4, 6, 8]) #공부시간
x2 = np.array([1, 2, 1, 3]) # 문제집 푼 갯수
y = np.array([20, 55, 45, 70]) # 점수


a1 = 0
a2 = 0
b = 0
lr = 0.005 #learning rate

for n in range(0,3000):
    y_hat = a1*x1 + a2*x2 + b
    error = y_hat - y
    a1_diff = sum(2 * error * x1)
    a2_diff = sum(2 * error * x2)
    b_diff = sum(2 * error * 1)
    
    a1 = a1 - a1_diff * lr
    a2 = a2 - a2_diff * lr
    b = b - b_diff * lr
    
    #print(f"a1_diff: {a1_diff:.2f}, a2_diff: {a2_diff:.2f}, b_diff: {b_diff:.2f}")
```

```python
plt.scatter(x1,y)
plt.plot(x1, a1*x1+ a2*x2 + b, 'red')
plt.show()
```

```python
x = np.arange(-3, 3, 0.01)
y = np.sin(x)
plt.plot(x,y)


# sin 함수와 가장 오차가 적은 1차 방정식 찾기
import numpy as np

a1 = 0
b = 0
lr = 0.00001 #learning rate

for n in range(0,3000):
    y_hat = a1*x + b
    error = y_hat - y
    a1_diff = sum(2 * error * x)
    b_diff = sum(2 * error * 1)
    
    a1 = a1 - a1_diff * lr
    b = b - b_diff * lr
    #print(f"a1:{a1:.2f} ,b:{b:.2f}")
    
plt.plot(x, a1 * x +  b, 'red')
```

```python
x = np.arange(-3, 3, 0.01)
y = np.sin(x)
plt.plot(x,y)


# sin 함수와 가장 오차가 적은 3차 선 그래프 찾기
import numpy as np

a = 0
b = 0
c = 0
d = 0
lr = 0.00001 #learning rate

for n in range(0,3000):
    y_hat = a * x**3 + b * x**2 + c * x + d
    error = y_hat - y
    a_diff = sum(2 * error * x**3)
    b_diff = sum(2 * error * x**2)
    c_diff = sum(2 * error * x)
    d_diff = sum(2 * error * 1)
    
    a = a - a_diff * lr
    b = b - b_diff * lr
    c = c - c_diff * lr
    d = d - d_diff * lr
    #print(f"a1:{a1:.2f} ,b:{b:.2f}")
print(a,b,c,d)    
plt.plot(x, a*x**3 + b*x**2 + c*x + d, 'red')
```

```python
# ChatGPT는 신이야
import matplotlib.pyplot as plt

# 데이터 생성
x = np.arange(-3, 3, 0.01)
y = np.sin(x)

# 3차 다항식으로 근사
coefficients = np.polyfit(x, y, 3)
p = np.poly1d(coefficients)

# 원래 sin 함수와 근사한 2차 다항식 그래프
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)')
plt.plot(x, p(x), label='Quadratic Approximation')
plt.title('Sin(x) and its Quadratic Approximation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
###################
# 시그모이드 / 분류모델 : 어떤 값을 넣으면 0~1 사이의 값으로 반환
##################
def sigmoid(x):
    value = 1 / (1+np.exp(-x))
    return value

x = np.arange(-5, 5, 0.1)
y = sigmoid(x)

plt.plot(x,y)
plt.show()
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
import pandas as pd
import time

# WebDriver 경로 설정
driver = webdriver.Chrome()

# 다음 뉴스 검색 페이지로 이동
driver.get("https://search.daum.net/search?w=news&q=신한은+행")
```

```python
# 결과 저장을 위한 리스트
titles = []
summaries = []
df = []

# 뉴스 기사 수집
while len(titles) < 100:
    news_items = driver.find_elements(By.CSS_SELECTOR,'ul.c-list-basic li div.item-bundle-mid')
    for i in news_items:
        title = i.find_elements(By.CSS_SELECTOR, "div.item-title")
        for j in title:
            titles.append(j.text)
    
        summary = i.find_elements(By.CSS_SELECTOR, 'p.conts-desc.clamp-g2')
        for j in summary:
            summaries.append(j.text)
        
        if len(titles) >= 100:
            break
    else:
        # 다음 페이지로 이동
        next_button = driver.find_element(By.CSS_SELECTOR,'div.compo-paging.ty_research button.btn_next')
        next_button.click()
        time.sleep(1)  # 페이지 로딩 대기

# WebDriver 종료
driver.quit()

# 데이터프레임 생성
df = pd.DataFrame({'Title': titles, 'Summary': summaries})

# '신한', '신한은행' 단어 제거
df['Title'] = df['Title'].str.replace('신한은행', '').str.replace('신한', '')
df['Summary'] = df['Summary'].str.replace('신한은행', '').str.replace('신한', '')
```

```python
from konlpy.tag import Kkma #형태소분석기
import pandas as pd

kkma = Kkma() #형태소분석기 kkma 정의

df['SUM'] = df['Title'] + " " + df['Summary']
total = []

for i in df['SUM']:
    nouns = kkma.nouns(i)
    total+=nouns
    
from collections import Counter 
dic=Counter(total) # 제목/내용에 있는 명사(total)를 빈도를 센 것을 dic에 넣음

from wordcloud import WordCloud
import matplotlib.pyplot as plt

from PIL import Image
import numpy as np
image = Image.open("heart.png") # 하트 이미지를 가져옴
image2 = np.array(image) #이미지를 행렬로 바꿔라

wc = WordCloud(background_color="white",
              font_path="BMDOHYEON_otf.otf",
              colormap="Blues",
              mask = image2)\
            .generate_from_frequencies(dic)

plt.axis('off')
plt.imshow(wc)
plt.show()
```
