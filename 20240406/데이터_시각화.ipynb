## 데이터 시각화
데이터 시각화는 데이터 분석 과정에서 중요한 단계 중 하나로, 데이터를 그래픽 형태로 표현하는 과정을 의미합니다. 이는 숫자나 텍스트로만 이루어진 복잡한 데이터 세트를 이해하기 쉽고, 직관적인 시각적 형태로 전환하여, 패턴, 추세, 이상치 같은 중요 정보를 빠르게 식별할 수 있게 해줍니다.

````python
import matplotlib.pyplot as plt # Python 시각화 기능 기반
import seaborn as sns # 자동으로 예쁜 그림
import pandas as pd # 
````

````python
anscombe = sns.load_dataset("anscombe")
anscombe
data1 = anscombe[anscombe['dataset'] == 'I']
data2 = anscombe[anscombe['dataset'] == 'II']
data3 = anscombe[anscombe['dataset'] == 'III']
data4 = anscombe[anscombe['dataset'] == 'IV']
````

````python
print(data1['x'].mean(), data1['y'].mean())
print(data2['x'].mean(), data2['y'].mean())
print(data3['x'].mean(), data3['y'].mean())
print(data4['x'].mean(), data4['y'].mean())
print("##############################")
print(data1['x'].std(), data1['y'].std())
print(data2['x'].std(), data2['y'].std())
print(data3['x'].std(), data3['y'].std())
print(data4['x'].std(), data4['y'].std())
print("##############################")
print(data1['x'].corr(data1['y']))
print(data2['x'].corr(data2['y']))
print(data3['x'].corr(data3['y']))
print(data4['x'].corr(data4['y']))
````

````python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1) # 2행 2열의 1번째 자리
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.scatter(data1['x'], data1['y']) # 점그래프(산점도)
ax2.scatter(data2['x'], data2['y'])
ax3.scatter(data3['x'], data3['y'])
ax4.scatter(data4['x'], data4['y'])

ax1.set_title("data1")
ax2.set_title("data2")
ax3.set_title("data3")
ax4.set_title("data4")

fig.suptitle("Anscombe Data")

plt.tight_layout()
plt.savefig("Anscombe Data.png")
````

````python
import seaborn as sns

tips = sns.load_dataset("tips")
tips

# 라인그래프
#sns.lineplot(tips)

# 연속형 데이터 & 한개의 특성 : 히스토그램을 활용
sns.set_style('whitegrid')
sns.histplot(tips, x='total_bill', bins=20, kde=True)
````

````python
#2개의 연속형 데이터 : 산점도(이변량 그래프)
sns.lmplot(tips, x='total_bill', y='tip', hue='smoker')
````

````python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker', fit_reg=True)

sns.kdeplot(data=tips, x='total_bill', y='tip')
````

````python
# 히스토그램 + 산점도
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
````

````python
sns.boxplot(data=tips, x='sex', y='tip', hue='day')
````

````python
sns.pairplot(tips, hue='sex')
````

````python
#sns.kdeplot(bike, x='yr', y='cnt', hue='mnth')
sns.lineplot(bike, x='mnth', y='cnt', hue='yr')
````

````python
# 히트맵( 행:월, 열:cnt)
# pivot_table : 내가 원하는 행과 열을 설정해서 집계값
heatdata = bike.pivot_table(index="mnth", columns='weekday', values='cnt', aggfunc="sum") 
heatdata

#형을 월로, 열은 요일이 되었을 때, 각 값을 sum 한 결과
import seaborn as sns
sns.heatmap(heatdata, cmap="Blues", annot=True, fmt="d")
````
