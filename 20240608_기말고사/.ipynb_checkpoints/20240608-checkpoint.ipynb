```python
# 게임기 10대
class Game:
    def __init__(self, arms=10):
        self.rates = np.random.rand(arms)

    def play(self,arm):
        rate = self.rates[arm]
        if rate > np.random.rand():
            return 1 
        else: return 0
        
game = Game()
game.rates
```

```python
for i in range(10):
    a = game.play(1)
    print(a)
```

```python
Qs = np.zeros(10) # 게임기 확률
Ns = np.zeros(10) # 게임기를 이용한 횟수
```

```python
total_reward = []
for n in range(10000):
    action = np.random.randint(0, 10) # 0~9까지 정수 하나 랜덤으로 고름
    reward = game.play(action)
    
    total_reward.append(reward)
    
    Ns[action] += 1
    Qs[action] = (reward - Qs[action]) / Ns[action] + Qs[action]
```

```python
print(Qs)
print(game.rates)
```
