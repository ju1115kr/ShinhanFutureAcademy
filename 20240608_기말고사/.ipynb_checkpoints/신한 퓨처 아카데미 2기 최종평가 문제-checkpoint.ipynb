## 1. 문제에 제시된 페이지에 있는 내용을 수집해서 아래와 같이 출력하세요.
(URL: https://wikidocs.net/book/5011)

* 전혀 구현하지 못했다. (0점)
* 구현했지만 출력 결과와 동일하지 않다. (15점)
* 완벽히 구현하였다. (25점)

```python
# 웹 크롤링 (Web Crawling) - BeautifulSoup
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 웹 페이지 가져오기
url="https://wikidocs.net/book/5011"
response = requests.get(url)
html_content = response.content

# BeautifulSoup으로 HTML 파싱
soup = BeautifulSoup(html_content, "html.parser")
```

```python
list_group = soup.find("div", {'class': "list-group list-group-toc"})
list_items = list_group.find_all("a", {'class': "list-group-item"})

list = []
for list_item in list_items:
    list.append(list_item.text.strip())
df = pd.DataFrame(list)
df = df[:-1]
df["제목"] = df[0]

result = pd.DataFrame(df["제목"])
result
```

## 2. wine을 분류하는 머신러닝 모델을 만들고 학습한 뒤, 성능을 평가하고 모델을 해석하세요.
* 데이터 분할을 75:25 비율로 하세요.
* 데이터 표준화를 수행하세요.
* K 최근접 이웃 모델, 랜덤포레스트, XGBoost 모델을 만들고 어떤 모델이 성능이 더 좋은지 확인 후 제일 좋은 모델과 그 이유를 주석에 적어주세요.
* 모델의 성능 평가는 정확도, 정밀도, 재현율을 통해서 해석하세요.

### wine 데이터 특성 설명
이 데이터는 와인 품질 분석을 위한 데이터입니다.  

    * alcohol: 와인의 알코올 함량
    * malic_acid: 말산 함량
    * ash: 재의 양(와인을 증발시킨 후 남은 고형물의 양)
    * alcalinity_of_ash: 재의 알칼리도
    * magnesium: 마그네슘 함량
    * total_phenols: 총 페놀 함량
    * flavanoids: 플라보노이드 함량
    * nonflavanoid_phenols: 비플라보노이드 페놀 함량
    * proanthocyanins: 프로안토시아니딘 함량
    * color_intensity: 색의 강도
    * hue: 색조
    * od280/od315_of_diluted_wines: 희석된 와인의 흡광도 비율
    * proline: 프롤린 함량
    * target: 목표 변수
    
* 전혀 구현하지 못했다. (0점)
* 완벽히 모델들을 구현하지 못하였거나 해석이 부정확하다. (15점)
* 완벽히 모델들을 구현하고 정확하게 해석하였다. (25점)

```python
from sklearn.datasets import load_wine
import pandas as pd

wine = load_wine()

data = wine.data
target = wine.target

wine = pd.DataFrame(data, columns = wine['feature_names'])
wine['target'] = target
wine
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve, mean_squared_error, mean_absolute_error
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor 
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier 
from xgboost import XGBClassifier, plot_importance
```

```python
# 분할 및 표준화
wine = wine.dropna()

# 특성과 타겟 분리
X = wine.drop('target', axis=1).values
y = wine['target'].values

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 데이터 스케일링
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#### K 근접 이웃모델
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_scaled, y_train)
knn.score(X_test_scaled, y_test)
y_pred = knn.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"KNN score: {knn.score(X_test_scaled, y_test)}")
print(f"KNN Accuracy: {accuracy}")

#정확도, 정밀도, 재현율
print(classification_report(y_test, y_pred))
```

```python
#### 랜덤포레스트
rf = RandomForestClassifier(n_estimators=100, random_state=42, oob_score=True)
rf.fit(X_train, y_train) # 스케일링 필요없음
rf.oob_score_
y_pred = rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Random Forest Accuracy: {accuracy}")

#정확도, 정밀도, 재현율
print(classification_report(y_test, y_pred))

# 중요 특징 시각화
feature_importances = rf.feature_importances_
features = wine.drop('target', axis=1).columns.tolist()
plt.figure(figsize=(10, 6))
plt.barh(features, feature_importances)
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importances in Random Forest Model')
plt.show()
```

```python
#### XGBoost
# XGBoost 분류기 모델 정의 및 학습
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
y_pred = xgb.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"XGBoost Accuracy: {accuracy}")

#정확도, 정밀도, 재현율
print(classification_report(y_test, y_pred))

# 특성 중요도 시각화
feature_importances = xgb.feature_importances_
features = wine.drop('target', axis=1).columns.tolist()

plt.figure(figsize=(10, 6))
plt.barh(features, feature_importances)
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Feature Importances in XGBoost Classifier')
plt.show()
```

```python
################# 결론 #################
# K 최근접 이웃 모델, 랜덤포레스트, XGBoost 모델을 만들고 어떤 모델이 성능이 더 좋은지 확인 후 제일 좋은 모델과 그 이유를 주석에 적어주세요.
# 모델의 성능 평가는 정확도, 정밀도, 재현율을 통해서 해석하세요.

# Knn, rf, xgboost 중 가장 성능과 정확도가 높은 것은 rf(랜덤포레스트)였습니다.
# 성능평가 지표룰 보면 알수있다시피,rf의 정확도, 정밀도, 재현율, f1score 모두 1.0으로 100% 만족했고,
# feature importances도 적절하게 분류하였습니다.
#######################################
```

### 캘리포니아 주택 가격 데이터 특성 설명
이 데이터셋은 캘리포니아 주택 가격 예측을 위한 데이터셋 입니다.

    * MedInc: Median Income. 각 지역(블록 그룹)의 중간 소득
    * HouseAge: House Age. 각 지역의 주택의 중간 연령
    * AveRooms: Average Rooms. 각 주택당 평균 방 개수
    * AveBedrms: Average Bedrooms. 각 주택당 평균 침실 개수
    * Population: Population. 각 지역의 인구 수
    * AveOccup: Average Occupancy. 각 주택당 평균 거주자 수
    * Latitude: Latitude. 각 지역의 위도 정보
    * Longitude: Longitude. 각 지역의 경도 정보
    * target: Target. 타겟 변수로, 각 지역의 주택 가격의 중간값
    
* 전혀 구현하지 못했다. (0점)
* 완벽히 모델들을 구현하지 못하였거나 해석이 부정확하다. (15점)
* 완벽히 모델들을 구현하고 정확하게 해석하였다. (25점)

```python
from sklearn.datasets import fetch_california_housing

california = fetch_california_housing()

data = california.data
target = california.target

california = pd.DataFrame(data, columns = california['feature_names'])
california['target'] = target
california
```

```python
# 데이터 분할 및 표준화
# 특성과 타겟 분리
X = california.drop('target', axis=1).values
y = california['target'].values

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# 데이터 스케일링
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# K 근접 이웃 모델
knr = KNeighborsRegressor(n_neighbors=3)
knr.fit(X_train_scaled, y_train)
knr.score(X_test_scaled, y_test)
y_pred = knr.predict(X_test_scaled)

# MAE, RMSE, MSE, R^2 값
knr_mae = mean_absolute_error(y_test, y_pred)
knr_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
knr_mse = mean_squared_error(y_test, y_pred)
knr_r2 = r2_score(y_test, y_pred)

print(f"knr_mae: {knr_mae}")
print(f"knr_rmse: {knr_rmse}")
print(f"knr_mse: {knr_mse}")
print(f"knr_r2: {knr_r2}")
```

```python
# 랜덤포레스트 모델
rfr = RandomForestRegressor(n_estimators=100, random_state=42, oob_score=True)
rfr.fit(X_train, y_train) # 스케일링 필요없음
rfr.oob_score_
y_pred = rfr.predict(X_test)

# MAE, RMSE, MSE, R^2 값
rfr_mae = mean_absolute_error(y_test, y_pred)
rfr_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rfr_mse = mean_squared_error(y_test, y_pred)
rfr_r2 = r2_score(y_test, y_pred)

print(f"rfr_mae: {rfr_mae}")
print(f"rfr_rmse: {rfr_rmse}")
print(f"rfr_mse: {rfr_mse}")
print(f"rfr_r2: {rfr_r2}")

# 중요 특징 시각화
feature_importances = rfr.feature_importances_
features = X = california.drop('target', axis=1).columns.tolist()
plt.figure(figsize=(10, 6))
plt.barh(features, feature_importances)
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importances in Random Forest Regressor Model')
plt.show()
```

```python
# 선형회귀모델
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)

# MAE, RMSE, MSE, R^2 값
lr_mae = mean_absolute_error(y_test, y_pred)
lr_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
lr_mse = mean_squared_error(y_test, y_pred)
lr_r2 = r2_score(y_test, y_pred)

print(f"lr_mae: {lr_mae}")
print(f"lr_rmse: {lr_rmse}")
print(f"lr_mse: {lr_mse}")
print(f"lr_r2: {lr_r2}")
```

```python
################# 결론 #################
# K 최근접 이웃 모델, 랜덤포레스트 모델, 선형 회귀 모델을 만들고 어떤 모델이 성능이 더 좋은지 확인 후 제일 좋은 모델과 그 이유를 주석에 적어주세요.
# 성능평가는 MAE, RMSE, MSE, R^2 값을 통해서 해석하세요.

# Knr, rfr, lr 중 가장 성능과 정확도(r2:설명력, mae&rmse:오차)가 높은 것은 rfr(랜덤포레스트)였습니다.
# 성능평가 지표룰 보면 알수있다시피,rf의 mae, rmse, mse, r2를 볼 때 가장 좋은 값으로 만족했고,
# feature importances도 적절하게 분류하였습니다.
#######################################
```

## 4. iris 데이터의 최적의 군집이 몇 개인지 구하세요.
KMeans 알고리즘을 사용해서 최적의 군집이 몇 개인지 구하고 그 이유를 주석으로 적어주세요.

* 전혀 구현하지 못했다. (0점)
* 완벽히 현하지 못하였거나 해석이 부정확하다. (15점)
* 완벽히 구현하고 정확하게 해석하였다. (25점)

```python
from sklearn.datasets import load_iris

iris = load_iris()
iris = pd.DataFrame(iris.data, columns = iris['feature_names'])
iris
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Iris 데이터 로드
iris = load_iris()
X = iris.data

# 군집계산을 위한 빈 리스트 
cluster = []

# 군집 수를 1에서 10까지 변화시키면서 KMeans 알고리즘을 적용
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X)
    cluster.append(kmeans.inertia_)  # 값을 리스트에 추가

# 군집 값을 그래프로 시각화하여 가장 꺾이는 점 찾기
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), cluster, marker='o', linestyle='--')
plt.title('iris Cluster_KMeans')
plt.xlabel('Number of Clusters')
plt.ylabel('cluster')
plt.show()
```

```python
################# 결론 #################
# KMeans 알고리즘을 사용해서 iris 데이터의 최적의 군집이 몇 개인지 구하는 코드를 작성하고 그 이유를 주석으로 적어주세요.

# KMeans를 통해 군집 내 군집 중심 간의 거리를 구했습니다.
# for 1~10까지 n_cluster를 변화하면서 적용했고,
# 2~3에서 가장 많이 그래프가 꺾이는 것을 볼 수 있습니다.
# 따라서, 군집은 2~3개 사이이며, 이 중 군집이 제일 많이 꺾이게 되었다가 완만해지는 쪽이 그래프가 값 중 3이므로, 약 3개의 군집을 가지고 있음을 알 수 있습니다.
#######################################
```
